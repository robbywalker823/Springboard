```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('sqlite_db_pythonsqlite.db')
```

### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! 

```python
q10_query = """
SELECT * 
FROM (
      SELECT sub.facility, SUM(sub.cost) AS total_revenue
        FROM (
              SELECT f.name AS facility, 
                     CASE WHEN b.memid = 0
                          THEN f.guestcost * b.slots
                          ELSE f.membercost * b.slots
                     END AS cost
              FROM Bookings AS b
              INNER JOIN Facilities AS f 
              ON b.facid = f.facid
              INNER JOIN Members AS m 
              ON b.memid = m.memid
              )sub
GROUP BY sub.facility
)sub2
WHERE sub2.total_revenue < 1000
"""
```

```python
q10_df = pd.read_sql_query(q10_query, conn)
q10_df
```

### Q11: Produce a report of members and who recommended them in alphabetic surname, firstname order

```python
q11_query = """
SELECT m1.surname || ', ' || m1.firstname AS member, m2.surname || ', ' || m2.firstname AS recommended_by
FROM Members AS m1
LEFT JOIN Members AS m2
ON m1.recommendedby = m2.memid
AND m1.recommendedby <> 0
WHERE m1.memid <> 0
ORDER BY m1.surname
"""
```

```python
q11_df = pd.read_sql_query(q11_query, conn)
q11_df
```

### Q12: Find the facilities with their usage by member, but not guests

```python
q12_query = """
SELECT f.name AS facility, COUNT(bookid) AS member_bookings 
FROM Facilities AS f
INNER JOIN Bookings AS b
ON f.facid = b.facid
WHERE memid <> 0
GROUP BY b.facid
"""
```

```python
q12_df = pd.read_sql_query(q12_query, conn)
q12_df
```

### Q13: Find the facilities usage by month, but not guests

```python
q13_query = """
SELECT strftime('%m', starttime) AS month, Count(bookid) AS member_bookings 
FROM Bookings AS b
INNER JOIN Members AS m
ON b.memid = m.memid
WHERE b.memid <> 0
group by month
"""
```

```python
q13_df = pd.read_sql_query(q13_query, conn)
q13_df
```
