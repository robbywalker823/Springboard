```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create DataFrame from csv file
bites = pd.read_csv('Health_AnimalBites.csv')
bites.head()
```

```python
# Check how much data is missing from each column. Quite a lot for a number of columns
missing = pd.concat([bites.isnull().sum(), 100 * bites.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='%', ascending=False)
```

```python
# Select the columns we want to focus on
df = bites.loc[:, ['bite_date', 'SpeciesIDDesc', 'BreedIDDesc', 'GenderIDDesc', 'victim_zip', 'WhereBittenIDDesc']]
df.head()
```

```python
# Check datatypes
df.dtypes
```

```python
# Convert date column to datetime format
df.bite_date = pd.to_datetime(df.bite_date, errors = 'coerce')
df.dtypes
```

```python
# Line graph of bites by year
sns.set(font_scale=1.2)
by_year = df.groupby(df['bite_date'].dt.year)['bite_date'].count()
by_year.plot(figsize=(10, 7), xlim=(1985, 2017))
plt.show()
```

```python
# Bite counts by species and year
by_spec = df[['bite_date', 'SpeciesIDDesc']].pivot_table(index='bite_date', columns='SpeciesIDDesc', 
                                                              aggfunc=len, fill_value=0)
by_year_spec = by_spec.groupby(by_spec.index.year).sum()

by_year_spec.head()
```

```python
# Area plot showing bites per year by species
by_year_spec.plot.area(figsize=(10,7), xlim=(1985, 2017))
plt.title('Distribution of Species by Year')
plt.ylabel('Number of Bites')
plt.xlabel('Year')
plt.show()
```

```python
# Area plot showing bites per year by species for 2010-2017
by_year_spec.plot.area(figsize=(10,7), xlim=(2010, 2017))
plt.title('Distribution of Species by Year')
plt.ylabel('Number of Bites')
plt.xlabel('Year')
plt.show()
```

```python
# Bite count by species
species = df.SpeciesIDDesc
species = species.dropna()
species.value_counts()
```

```python
# Bar plot of bites by species
species = df.SpeciesIDDesc.value_counts()
plt.figure(figsize=(10,7))
ax = sns.barplot(x=species.index, y=species)
plt.xticks(rotation=70)
plt.ylabel("Number of Bites")
plt.title('Animal Bites by Species')
plt.show()
```

```python
# Bar plot of bites by species without dogs or cats
plt.figure(figsize=(10,7))
ax = sns.barplot(x=species[2:].index, y=species[2:])
plt.xticks(rotation=70)
plt.ylabel("Number of Bites")
plt.title('Animal Bites Exluding Dogs and Cats')
plt.show()
```

```python
# Bar plot of dog bites by breed
breed = df.BreedIDDesc.value_counts()
plt.figure(figsize=(10,7))
ax = sns.barplot(x=breed[0:10].index,y=breed[0:10])
plt.xticks(rotation=70)
plt.ylabel("Number of Bites")
plt.title('Top Ten Highest Dog Bites by Breed')
plt.show()
```

```python
# Bar plot of bites by month
by_month = df.groupby(df['bite_date'].dt.month)['bite_date'].count()
plt.figure(figsize=(10,7))
ax = sns.barplot(x=by_month.index, y=by_month, palette='Blues_d')
plt.xlabel('Month')
plt.ylabel('Number of Bites')
plt.title('Total Number of Bites by Month')
plt.show()
```

```python
# Bite count per month by species
by_month_spec = by_spec.groupby(by_spec.index.month).sum()
by_month_spec
```

```python
# Bites per month by species
by_month_spec.plot(kind='bar', figsize=(15, 10), stacked=True)
plt.xlabel('Month')
plt.ylabel('Number of Bites')
plt.title('Total Bites by Month')
plt.show()
```

```python
# Horizonal bar chart showing where on body bite occurred
df['WhereBittenIDDesc'].value_counts().plot(kind='barh')
plt.title('Where Bites Ocurred')
plt.show()
```

```python
# Horizonal bar chart showing gender of animal
df['GenderIDDesc'].value_counts().plot(kind='barh')
plt.title('Gender of Animal')
plt.show()
```

```python
# Library with information on US Zipcodes
from uszipcode import SearchEngine
```

```python
search = SearchEngine()
```

```python
# The data comes from 234 separate zipcodes
len(df['victim_zip'].unique())
```

```python
# Bites per zipcode
df['victim_zip'].value_counts()[:40].plot(kind='bar', figsize=(15, 10))
plt.title('Zip Code of Victim')
plt.xticks(rotation=70)
plt.show()
```

```python
# Information on zipcode with most bites
zipcode = search.by_zipcode('40272')
zipcode
```

```python
# Returns state of zipcode
def zipcode_state(zipcode):
    return search.by_zipcode(zipcode).state
```

```python
# Returns county of zipcode
def zipcode_county(zipcode):
    return search.by_zipcode(zipcode).county
```

```python
# Create new columns for state and county of bites
df['county'] = df['victim_zip'].apply(lambda x: zipcode_county(x))
df['state'] = df['victim_zip'].apply(lambda x: zipcode_state(x))

df.head()
```

```python
# Bite count by county
df['county'].value_counts()
```

```python
# Bar chart of bites by county
df['county'].value_counts()[:10].plot(kind='bar', figsize=(10, 7))
plt.xticks(rotation=70)
plt.title('Bites by County')
plt.show()
```

```python
# Bite count by state
df['state'].value_counts()
```

```python
# Bar chart of bites by state
df['state'].value_counts()[:10].plot(kind='bar', figsize=(10, 7))
plt.xticks(rotation=0)
plt.title('Bites by State')
plt.show()
```

```python
# Returns population of zipcode
def zipcode_pop(zipcode):
    return search.by_zipcode(zipcode).population
```

```python
# Returns median income of zipcode
def zipcode_income(zipcode):
    return search.by_zipcode(zipcode).median_household_income
```

```python
# Create DataFrame of bite count by zipcode
zips = df['victim_zip'].value_counts()
zips = zips.to_frame()
zips.rename(columns={'victim_zip': 'bite_count'}, inplace=True)
zips.head()
```

```python
# Add columns for population and income
zips['population'] = zips.index
zips['population'] = zips['population'].apply(lambda x: zipcode_pop(x))

zips['income'] = zips.index
zips['income'] = zips['income'].apply(lambda x: zipcode_income(x))

zips.head()
```

```python
# No strong relationship between bite count and population or income
zips.corr()
```

```python
# Bubble plot showing population, income, and bite count
X = zips['population']
Y = zips['income']
S = zips['bite_count']
plt.figure(figsize=(12, 8))
plt.scatter(X, Y, S*5, alpha=0.5)
plt.xlabel('Population')
plt.ylabel('Median Income')
plt.title('Zip Code Demographics (Dot Size = Number of Bites)')
plt.show()
```
